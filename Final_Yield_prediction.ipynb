```python
import pandas as pd
```

```python
data =pd.read_csv('crop_yield.csv')
```

```python
data.head()
data = data[:10000]
```

```python
data.head()
```

```python
data.info()
```

```python
#handling missing data
data = data.dropna() #dropna()-to analyze and drop Rows/Columns with Null values. 
data.info()
```

```python
data.describe()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

Crop = le.fit_transform(data.Crop)
Crop_Year = le.fit_transform(data.Crop_Year)
Season = le.fit_transform(data.Season)
State = le.fit_transform(data.State)

data['Crop'] = Crop
data['Crop_Year'] = Crop_Year
data['Season'] = Season
data['State'] = State
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split
#from sklearn.cross_validation import train_test_split
```

```python
X = data.iloc[:,:-1]
y = data.iloc[:,-1]
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor


from sklearn.metrics import roc_auc_score , classification_report, mean_squared_error, r2_score


algorithms = [RandomForestRegressor(), LinearRegression(), KNeighborsRegressor(), SVR(), DecisionTreeRegressor()]
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,y,test_size=0.2,random_state=100)
```

```python
algorithms = [RandomForestRegressor(), LinearRegression(), KNeighborsRegressor(), SVR(), DecisionTreeRegressor()]
```

```python

# Create a dictionary to store the results
results = {}

# Iterate over the algorithms
for algorithm in algorithms:

    # Train the algorithm
    algorithm.fit(X_train, Y_train)

    # Make predictions on the test set
    #y_pred = algorithm.predict(X_test)
    y_train_pred = algorithm.predict(X_train)
    y_test_pred = algorithm.predict(X_test)

    # Calculate the MSE and R^2 scores
    mse = mean_squared_error(Y_test, y_test_pred)
    r2 = r2_score(Y_test, y_test_pred)

    # Store the results in the dictionary
    results[algorithm.__class__.__name__] = [mse, r2]
```

```python
# Print the results in a table
print('Algorithm | MSE | R^2')
print('----------|-----|-----')
for algorithm, result in results.items():
    print(f'{algorithm} | {result[0]:.3f} | {result[1]:.3f}')
  
```

```python

forest = RandomForestRegressor(n_estimators=1000, 
                               criterion='squared_error', 
                               random_state=1, 
                               n_jobs=-1)
forest.fit(X_train,Y_train)
y_train_pred = forest.predict(X_train)
y_test_pred = forest.predict(X_test)

print('MSE train: %.3f, test: %.3f' % (
        mean_squared_error(Y_train, y_train_pred),
        mean_squared_error(Y_test, y_test_pred)))
print('R^2 train: %.3f, test: %.3f' % (
        r2_score(Y_train, y_train_pred),
        r2_score(Y_test, y_test_pred)))
```
